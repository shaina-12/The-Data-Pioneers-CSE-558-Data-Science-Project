```python
import gzip
import json
import pandas as pd
```

```python
def convertJson(input_file, output_file):
  data = []
  with gzip.open(input_file, 'rt', encoding='utf-8') as fin:
    for line in fin:
      data.append(json.loads(line))
  with open(output_file, 'w', encoding='utf-8') as fout:
    json.dump(data, fout, indent=4)


def convertJsonToPickle(input_file, output_file):
  df = pd.read_json(input_file)
  df.to_pickle(output_file)

def convertPickleToCSV(input_file, output_file):
  df = pd.read_pickle(input_file)
  df.to_csv(output_file, index=False)
```

```python
convertJson('/content/drive/MyDrive/Data Science Project/Handmade_Products.jsonl.gz', '/content/drive/MyDrive/Data Science Project/Handmade_Products.json')
convertJson('/content/drive/MyDrive/Data Science Project/meta_Handmade_Products.jsonl.gz', '/content/drive/MyDrive/Data Science Project/meta_data_handmade.json')
```

```python
convertJsonToPickle('/content/drive/MyDrive/Data Science Project/meta_data_handmade.json', '/content/meta_data_handmade.pkl')
convertPickleToCSV('/content/meta_data_handmade.pkl', '/content/drive/MyDrive/Data Science Project/meta_data_handmade.csv')
```
