```python
text = list(df['processed_text'])
new_text = []
for i in range(len(text)):
  new_text.append(text[i].split(" "))
```

```python
documents = [TaggedDocument(words=doc, tags=[i]) for i, doc in enumerate(new_text)]
```

```python
model = Doc2Vec(documents, vector_size=100, window=3, min_count=1, workers=2)
```

```python
# vectors = []
# for i in range(len(new_text)):
#   vectors.append(model.infer_vector(new_text[i]))

vectors = [model.infer_vector(txt) for txt in new_text]
```

```python
vectors = np.array(vectors)
```

```python
vectors.shape
```

```python
df1['index'] = list(df1.index)
```

```python
indexes = list(df1['index'])
```

```python
doc_vec_dict = {"index":indexes, 'doc_to_vec':vectors}
```

```python
with open('/content/drive/MyDrive/Data Science Project/doc_to_vec_vectors.pkl', 'wb') as file:
    pickle.dump(doc_vec_dict, file)
```
